```python
# Beginning of the EDA analysis and visualization.
import pandas as pd
file_path = 'CVD_cleaned.csv'
data = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as mmsno
from collections import Counter
```

```python
# Viz missing data and clean dataset
plt.figure(figsize=(8,5))
sns.heatmap(data.isnull(), annot=True, cmap='coolwarm')
mmsno.bar(data, color='green')
plt.title("Visualization of Missing Data")
plt.show()
```

```python
# Individual count of the General Health types.
data["General_Health"].value_counts()
```

```python
#Viz for patients General health with specific counts per health type
ax = sns.countplot(x= data["General_Health"], order=data["General_Health"].value_counts(ascending=False).index)
values = data["General_Health"].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=values)
plt.title("General Health Status of Patients");
```

Over 110,000 patients reported very good health status while 95,364 reported Good health status. There were 55,954 patients with excellent health while 35,810 reported fair. A total of 11,331 patients had poor health.

```python
# Viz for the above data in pie chart to show us percentages of each health type
fig, ax = plt.subplots(figsize=(5,5))
count = Counter(data["General_Health"])
ax.pie(count.values(), labels=count.keys(), autopct=lambda p:f'{p:.1f}%')
ax.set_title("Percentage of Patient's Health Categories")
plt.show();
```

35.7% of CVD patients reported their health was very good, 30.9% reported theirs was good. 18.1% had an excellent health status while 11.6% reported fair and 3.7% had poor health.  

```python
#Viz for patients general health with specific counts per health type by checkup
plt.figure(figsize=(15,5))
ax = sns.countplot(y= data["Checkup"], order=data["Checkup"].value_counts(ascending=False).index)
values = data["Checkup"].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=values)
plt.title("Checkup Status of Patients");
```

Over 75% of patients did a CVD checkup regarding their health within the past one year while just 12% did a checkup within the past two years. Only 1,407 representing 0.5% have never done a checkup.

```python
# Visualization of patient health using Bivariate analysis
# General Health by BMI
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(14,5))
aa=data.groupby("General_Health")["BMI"].sum().reset_index()
sns.barplot(x='General_Health', data=aa, y='BMI', ax=axs[0])
axs[0].set_title("General Health by BMI");

# Visualization of patient health using Bivariate analysis
# General Health by Weight_(kg),
ab=data.groupby("General_Health")["Weight_(kg)"].sum().reset_index()
sns.barplot(x='General_Health', data=ab, y='Weight_(kg)', ax=axs[1])
axs[1].set_title("General Health by Weight(kg)");

# Visualization of patient health using Bivariate analysis
# General Health by Height_(cm)
ac=data.groupby("General_Health")["Height_(cm)"].sum().reset_index()
sns.barplot(x='General_Health', data=ac, y='Height_(cm)', ax=axs[2])
axs[2].set_title("General Health by Height(cm)");
```

There's a direct proportion between patient's general health and BMI, Weight and height. As patient health improves so also does these metrics increase and vice versa.

```python
# Viz data in pie chart to show Patient's health by Gender
fig, ax = plt.subplots(figsize=(5,5))
count = Counter(data["Sex"])
ax.pie(count.values(), labels=count.keys(), autopct=lambda p:f'{p:.1f}%')
ax.set_title("Percentage of Patient's Health by Gender")
plt.show();
```

Percentage of male and female CVD patients are almost the same with male patients examined at 52% while their female counterpart were at 48%.

```python
# Group patient age in groups
def age_group(x):
  if x == '18-24':
    return "(18-24):Youngster"
  elif x == '25-29':
    return "(25-29):Young Adult"
  elif x == '30-34':
    return "(30-34):Mid Adult"
  elif x == '35-39':
    return "(35-39):Adult"
  elif x == '40-44':
    return "(40-44):Older Adult"
  elif x == '45-49':
    return "(45-49):Senior Adult"
  elif x == '50-54':
    return "(50-54):Young Senior"
  elif x == '60-64':
    return "(60-64):Mid Senior"
  elif x == '65-69':
    return "(65-69):Old Senior"
  elif x == '70-74':
    return "(70-74):Elder"
  elif x == '75-79':
    return "(75-79):Mid Elder"
  else:
    return "(80+):Senior Elder"

#Apply function to the data
data["Patient_age_groups"] = data["Age_Category"].apply(age_group)
```

```python
#Viz for count by patient_age_groups
plt.figure(figsize=(15,5))
ax = sns.countplot(y= data["Patient_age_groups"], order=data["Patient_age_groups"].value_counts(ascending=False).index)
values = data["Patient_age_groups"].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=values)
plt.title("Count by Patient_age_groups");
```

More seniors upted for for the CDA examinations than young adults which is understandable because the older patients get the more conscious they become of their health and risk factors.

```python
# Viz to show General health of patients by Consumption types
consum_type = data.groupby("General_Health")[["Alcohol_Consumption", "Fruit_Consumption", "Green_Vegetables_Consumption", "FriedPotato_Consumption"]].sum().reset_index()
consum_type = pd.melt(consum_type, id_vars="General_Health", var_name="metric", value_name="Total")
sns.barplot(data=consum_type, x='General_Health', y='Total', hue="metric")
plt.title("General Health of CVD Patients by Consumption Types")
plt.show();
```

The Viz lucidly shows that the CVD patients with good health generally consume lots of fruits and green vegetables from their diet with moderate fried potato & alcohol intake while does with fair and poor health consume less of fruits, green vegetables & fried potato with insignificantly low alcohol intake.

```python
# function to digitize the Exercise column into two groups
def exercise_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Exercise_Group"] = data["Exercise"].apply(exercise_group)
```

```python
# function to digitize the Heart_Disease column into two groups
def heart_disease_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Heart_Disease_Group"] = data["Heart_Disease"].apply(heart_disease_group)
```

```python
# function to digitize the Skin Cancer column into two groups
def skin_cancer_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Skin_Cancer_Group"] = data["Skin_Cancer"].apply(skin_cancer_group)
```

```python
# function to digitize the Other Cancer column into two groups
def other_cancer_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Other_Cancer_Group"] = data["Other_Cancer"].apply(other_cancer_group)
```

```python
# function to digitize the Other Depression column into two groups
def depression_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Depression_Group"] = data["Depression"].apply(depression_group)
```

```python
# function to digitize the Other Diabetes column into two groups
def diabetes_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Diabetes_Group"] = data["Diabetes"].apply(diabetes_group)
```

```python
# function to digitize the Other Arthritis column into two groups
def arthritis_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Arthritis_Group"] = data["Arthritis"].apply(arthritis_group)
```

```python
# function to digitize the Other Smoking history column into two groups
def smoking_history_group(a):
  if a == 'Yes':
    return 1
  else:
    return 0

#Apply function to the data
data["Smoking_History_Group"] = data["Smoking_History"].apply(smoking_history_group)
```

```python
# Viz to show General health of patients by Disease Type
disease_type = data.groupby("General_Health")[["Exercise_Group", "Heart_Disease_Group", "Skin_Cancer_Group", "Other_Cancer_Group", "Depression_Group", "Diabetes_Group", "Arthritis_Group", "Smoking_History_Group"]].sum().reset_index()
disease_type = pd.melt(disease_type, id_vars="General_Health", var_name="metric", value_name="Total")
sns.barplot(data=disease_type, x='General_Health', y='Total', hue="metric")
plt.title("General Health of CVD Patients by Disease Types")
plt.show();
```

The chart simply illustrates the benefits exercise has on good health. CVD patients who have incorporated frequent or high exercise routines in their life style benefited tremendiously. Patients with very low exercise routine fell into the category of patients with poor health. Also, patients with good health recorded moderate to low heart disease and diabetes. Regarding mental health (depression), patients with excellent & good health recorded low to moderate depression while it was moderate to low for patients with fair and poor health. In general, patients with poor health recorded very low indicies in all metrics including exercise and mental health while patients with good and excellent health recorded moderate to high indicies on all metrics including exercise.

```python
# Viz showing Bivariate analysis
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
sns.boxplot(x="Depression", y="Alcohol_Consumption", data=data, ax=axs[0])
axs[0].set_title("Depression & Alcohol Consumption");

sns.boxplot(x="General_Health", y="Fruit_Consumption", data=data, ax=axs[1])
axs[1].set_title("General Health & Fruit Consumption");

sns.boxplot(x="Heart_Disease", y="Alcohol_Consumption", data=data, ax=axs[2])
axs[2].set_title("Heart Disease & Alcohol Consumption");

```

The first boxplot shows the relationship between two variables. Alcohol consumption on the y-axis and depression on the x-axis. The plot examines individual patients with and without depression against the quantity of alcohol consumed. Clearly, those with no recorded depression consumed slightly more alcohol with both groups having a mean of 2 consumptions per frequency and various degree of outliers.

The second boxplot shows the relationship between two important variables. Quanity of fruits consumed on the y-axis and quality of health on the x-axis. Patients in the excellent health group recorded the highest fruit intake and a mean of 30 consumptions per frequency with no outliers while patients in both poor and fair health groups recorded a moderate consumption with a mean of 20 per frequency with outliers. Those in the good and very good health groups also recorded moderate fruit consumptions with the good health category recording a mean consumption of 24 per frequecy and a handful of outliers.

The last boxplot explains the relationship between alcohol consumption & heart disease. Patients with no heart disease consumed more alcohol than those who confirmed they had a heart-related disease. While they were outliers for both groups, those with no heart-related disease recorded a mean value of 2 consumptions per frequency.
